# Backtracking

```python
def permutation(options: str) -> list:
    ans = []
    def solver(path: str, opts: str):
        if len(opts) == 0:
            ans.append(path)
        else:
            for i in range(len(opts)):
                rem = opts[0:i]+opts[i+1:]
                solver(path + opts[i], rem)
    solver("", options)
    return ans
```

```python
permutation("abcde")
```
